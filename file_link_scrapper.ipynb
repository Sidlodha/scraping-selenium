```python
options = Options()
options.headless = True
driver = webdriver.Firefox(options = options)
driver.set_window_size(1550,700)
links_list = []
left_links = []
with open("w_pants.txt", 'r') as f:
    i = 0
    for line in f:
        print(line,'initial')
        driver.get(line)
        try:
            driver.find_element(By.CSS_SELECTOR, ".desktop").click()
        except:
            pass
        new_link = line.split("?")
        if len(new_link)>1:
            new_link = new_link[0]
        else:
            new_link = new_link[0][:-1]
        new_links = []
        SCROLL_PAUSE_TIME = 3

        last_height = driver.execute_script("return document.body.scrollHeight")

        while True:
        # Scroll down to bottom
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait to load page
            time.sleep(SCROLL_PAUSE_TIME)

            # Calculate new scroll height and compare with last scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        colors=driver.find_elements_by_css_selector('div.color-container')
        
        for color in colors:
            color_id = None
            color_id = color.get_attribute('id')
            print(color_id,'color_id')
            new_link_color = new_link + "?c=" + color_id
            new_links.append(new_link_color)
        links_list.extend(new_links)
        i+=1
        colors = None
        print(i,'i')
            
    print(len(links_list))
    with open('pants_women_mango.txt', 'w') as f:
        for item in links_list:
            f.write("%s\n" % item)
    driver.quit()
```
