```python
def extract_link(file):
    os.chdir('/home/uplarauser/gymshark/')
    print(os.getcwd())
    f = open(file,'r')
    temp = []
    for line in f:
        line = line.replace('\n','')
        temp.append(line)
    temp = list(set(temp))
    print(len(temp))
    curr_dir = os.getcwd() + '/'
    os.chdir(curr_dir + file.replace('.txt','') + '/gymshark/')
    main_dir = os.getcwd()
    options = Options()
    options.headless = True
    driver = webdriver.Firefox(options=options)
    driver.maximize_window()
    i = 0
    SCROLL_PAUSE_TIME = 1 
    for link in temp:
        print(i,link)

        product_name = link.split('/')[4]
        print(product_name)
        if os.path.isdir(main_dir + "/" + product_name + "/"):
            print('present')
            continue
        driver.get(link)
        os.system("mkdir " + product_name)
        os.chdir(main_dir + "/"+ product_name + "/")
    #   Infinite scroll handling
        try:
            close = driver.find_element_by_css_selector('button.locale-btn')
            close.click()
        except:
            print('pass')
            pass
        colors = driver.find_elements_by_css_selector('div.pinfo__thumbnails__wrap > a')
        print(len(colors))
        imgs = driver.find_elements_by_css_selector('div.pimages > article')
        srcs = []
        for img in imgs:
            src = img.get_attribute('data-image')
            if src!=None:
                srcs.append(src)

        srcs = list(set(srcs))
        flag = 1
        for x in srcs:
            try:
                urllib.request.urlretrieve(x, str(flag) + ".jpg")
                flag+=1
            except:
                pass                
            print(x)
        print(len(srcs))
        product_colors = {}
        os.chdir(main_dir)
        i+=1
        print(i,'ends')
    driver.quit()
```
